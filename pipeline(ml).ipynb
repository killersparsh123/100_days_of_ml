```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2 
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline,make_pipeline

```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Survived'),df['Survived'],test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
#for removing the null values using imputer
trf1=ColumnTransformer(
    [('impute_age',SimpleImputer(),[2]),
     ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])],remainder="passthrough")
     
```

```python
#applying One Hot Encoding
trf2=ColumnTransformer(
    [('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1,6])],remainder='passthrough')
```

```python
#Scaling
trf3=ColumnTransformer(
    [('scale',MinMaxScaler(),slice(0,10))])
```

```python
#Feature Selection
trf4=SelectKBest(score_func=chi2,k=5)
```

```python
#train the model
trf5=DecisionTreeClassifier()
```

CREATE PIPELINE


```python
pipe=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)])
    
```

```python
pipe.fit(x_train,y_train)
```

Pipeline V/S make_pipeline


```python
#alternate syntax
pipe=make_pipeline(tr1,trf2,trf3,trf4,trf5)
#train 
pipe.fit(x_train,y_train)
```

Exporting Pipeline 


```python
#export
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
